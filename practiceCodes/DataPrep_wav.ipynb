```python
# Perform the reading through all audio files
for i in range(len(wav_audio_files)):
    # Getting audio and frequency in seconds of the first file
    audio, freq_sec = lr.load(wav_audio_files[i])
    # Getting the time array
    time = np.arange(len(audio))/freq_sec
    
    fig, ax = plt.subplots()
    ax.plot(time, audio)
    ax.set(xlabel="Time (s)", ylabel="Amplitude")
    plt.show()
```
